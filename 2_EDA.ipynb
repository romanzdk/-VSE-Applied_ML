```python
# define categorical and numerical columns
cat_c = ['Hráč', 'Oddíl', 'Družstvo', 'Soupeř', 'Oddíl soupeř','Družstvo soupeř', 'Soutěž']
num_c = []
```

# Value counts for categorical variables

```python
for c in cat_c:
    print(df[c].value_counts()[:5])
    print()
```

# Histograms

```python
for c in num_c:
    plt.hist(df[c])
    plt.title(c)
    plt.show()
```

# Correlation

```python
try:
    fig, ax = plt.subplots(figsize=(10,10))         # Sample figsize in inches
    sns.heatmap(df[num_c].corr())
except:
    pass
```
