```python
# define categorical and numerical columns
cat_c = df.columns
num_c = []
```

# Value counts for categorical variables

```python
for c in cat_c:
    try:
        print(f"{30*'-'}")
        print(df[c].value_counts()[:5])
        print()
    except:
        pass
```

# Histograms

```python
for c in num_c:
    plt.hist(df[c])
    plt.title(c)
    plt.show()
```

# Correlation

```python
try:
    fig, ax = plt.subplots(figsize=(10,10))         # Sample figsize in inches
    sns.heatmap(df[num_c].corr())
except:
    pass
```
